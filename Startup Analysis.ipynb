```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import squarify

startup_data = pd.read_csv('indian_startup_funding.csv') 
```

# Data Cleaning and Formatting

```python
df = pd.DataFrame(startup_data)
df.dropna()
df.head()
```

Percentage ( % ) of null values w.r.t. columns in the dataset:

```python
(df.isnull().sum() / df.shape[0] * 100).sort_values(ascending = False).round(2).astype(str) + ' %'
```

```python
df['Amount in USD'] = df['Amount in USD'].apply(lambda x : str(x).replace("nan", "0"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("ECommerce", "E-Commerce"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("E-commerce", "E-Commerce"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("eCommerce", "E-Commerce"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Ecommerce", "E-Commerce"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("E - Commerce", "E-Commerce"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Fin-Tech", "FinTech"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Fintech", "FinTech"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("EduTech", "EdTech"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Ed-Tech", "EdTech"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Edu-tech", "EdTech"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("EduTech", "EdTech"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Education", "EdTech"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Food and Beverage", "Food & Beverage"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("IT", "Technology"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Information Technology", "Technology"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace(" s", "Misc"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Food & Beverages", "Food & Beverage"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Food & Beverages", "Food & Beverage"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Transportationation", "Transport"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Health and Wellness","Healthcare"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("Transportation","Transport"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("EdTech","Education"))
df['Industry'] = df['Industry'].apply(lambda x : str(x).replace("nan","NA"))
df['Investors'] = df['Investors'].apply(lambda x : str(x).replace("undisclosed investors","Undisclosed Investors"))
df['Investors'] = df['Investors'].apply(lambda x : str(x).replace("Undisclosed Investor","Undisclosed Investors"))
df['Investors'] = df['Investors'].apply(lambda x : str(x).replace("Undisclosed investor","Undisclosed Investors"))
df['Investors'] = df['Investors'].apply(lambda x : str(x).replace("Undisclosed Investorss","Undisclosed Investors"))
df['Investors'] = df['Investors'].apply(lambda x : str(x).replace("nan","Unkown Sources"))
df['Description'] = df['Description'].apply(lambda x : str(x).replace("Online lending platform","Online Lending Platform"))
df['Description'] = df['Description'].apply(lambda x : str(x).replace("Online Lending platform","Online Lending Platform"))
df['Description'] = df['Description'].apply(lambda x : str(x).replace("Online learning platform","Online Learning Platform"))
df['Investment Type'] = df['Investment Type'].apply(lambda x : str(x).replace("Seed\nFunding","Seed"))
df['Location'] = df['Location'].apply(lambda x : str(x).replace("Bengaluru","Bangalore"))
df['Location'] = df['Location'].apply(lambda x : str(x).replace("Gurugram","Gurgaon"))
df['Location'] = df['Location'].apply(lambda x : str(x).replace("nan","Others"))
df['Investment Type'] = df['Investment Type'].apply(lambda x : str(x).replace("Seed/Angel","Seed / Angel"))
df['Investment Type'] = df['Investment Type'].apply(lambda x : str(x).replace("Seed/ Angel","Seed / Angel"))
df['Investment Type'] = df['Investment Type'].apply(lambda x : str(x).replace("Seed Funding","Seed"))
df['Investment Type'] = df['Investment Type'].apply(lambda x : str(x).replace("Angel / Seed","Seed / Angel"))
df['Investment Type'] = df['Investment Type'].apply(lambda x : str(x).replace("Seed Round","Seed"))
df['Investment Type'] = df['Investment Type'].apply(lambda x : str(x).replace("Seed / Angel Funding","Seed / Angel"))
df['Investment Type'] = df['Investment Type'].apply(lambda x : str(x).replace("nan","Private Equity"))

```

```python
(df.isnull().sum() / df.shape[0] * 100).sort_values(ascending = False).round(2).astype(str) + ' %'
```

```python
df['Amount in USD'] = pd.to_numeric(df['Amount in USD'])
del df["Website URL"]
df.head()
```

```python
df['Date'] = pd.to_datetime(df.Date, dayfirst = True)
df.head()
```

# Fundings Across The Years

```python
time_df = df
time_df["Year"] = (pd.to_datetime(time_df['Date']).dt.year)

temp = time_df['Year'].value_counts().head(10)

print("Number of funding per year in decreasing order(Top 10)\n",temp)
year_month = time_df['Year'].value_counts()
time_df['Year'].value_counts().plot(kind='pie', figsize=(11,7), title='Number of Startups over Years', autopct='%.f%%')
plt.show()
```

We can see that 2018 - 2021 had a huge decline in number of startups funded dropping from 687 in '17 to 310 in '18 and another drop in '19 to only 111 fundings

# Most Funded Startups and Startups with Most Number Of Rounds

```python
temp_df = df.groupby('Startup Name')["Amount in USD"].sum().reset_index()

temp_df = temp_df.sort_values('Amount in USD', ascending=False).head(10)

temp_df = temp_df.set_index('Startup Name', drop = True)

temp_df.plot(kind='pie', y = 'Amount in USD', subplots=True, figsize=(12,6), autopct='%.f%%')

plt.show()
```

### Distribution of Most Common Funding Stages

```python
df['Investment Type'].value_counts().head(10)
```

```python
investment_type = df['Investment Type'].value_counts().head(10)

plt.figure(figsize = (10, 7))
squarify.plot(sizes = investment_type.values, label = investment_type.index, value = investment_type.values)
plt.title('Investment type distribution', fontdict = {'fontname' : 'Monospace', 'fontsize' : 30, 'fontweight' : 'bold'})
plt.show()
```

The major types of investments are Private Equity and Seed Funding which account for more than 90% of fundings.

Other types of fundings are Debt funding, Seed/Angel funding, Series A, Series B, Series C but they are very rare.
